```python
#Noah Greski - Testing Data ML project

#imports
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
#read in data
housing_data = pd.read_csv(Path("housing data.csv"))
housing_data
```

```python
housing_data.head()
```

```python
housing_data.info()
```

```python
#check for missing values
housing_data.isnull().sum()
```

```python
#create dataframe with only numerical data
housing_num = housing_data.select_dtypes(include=[np.number])
housing_num.head()
```

```python
#visualize the data
g = sns.PairGrid(housing_num)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
#visual the housing prices
housing_data['price'].hist(bins =10)
plt.xlim = (0, 1000000)
housing_data.sort_values(by = 'price', ascending = False)
```
