```python
#create dataframe with only numerical data
housing_num = housing_data.drop(['country','city','street','date'], axis=1)
housing_num.head()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
# Price Distribution
plt.figure(figsize=(10, 6))
plt.hist(housing_num['price'], bins=20, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histogram of Prices')
plt.xlabel('Price (Tens of Millions)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Separate the houses by price over/under $5 million to isolate outliers
condition_value = 1000000  # Million dollar homes
condition_column = 'price'  # Sort by price

# Split based on condition
over = housing_num[housing_num['price'] >= condition_value]
under = housing_num[housing_num['price'] < condition_value]

print("Homes Over $1m: ", len(over), "\nHomes Under $1m: ", len(under))
```

```python
sns.scatterplot(data = under, x = under['sqft_living'], y = under['price'])
```

```python
under = under[under['price'] > 0]
sns.scatterplot(data = under, x = under['sqft_living'], y = under['price'])
```

```python
len(under) # Total points of training data
```

```python
housing_features = under.drop(columns=['price'])  # Features
housing_target = under['price']  # Target variable

#numerical features scaling

from sklearn.pipeline import Pipeline
num_pipeline = Pipeline([
    ("impute", SimpleImputer(strategy="median")),
    ("standardize", StandardScaler()),
])
#object feature transform

cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore"))

#A pipeline function for both

preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)),
    (cat_pipeline, make_column_selector(dtype_include=object)),
)

#Transform
housing_features_prepared = preprocessing.fit_transform(housing_features)
```

```python
# Revisualize for o/u $1m homes
plt.figure(figsize=(10, 6))
plt.hist(under['price'], bins=10, color='blue', edgecolor='black') # <$1m
plt.title('Histogram of Home Prices Under $1m')
plt.xlabel('Price (Millions of Dollars)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create and run a linear regression model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
# Step 2: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(housing_features_prepared, housing_target, test_size=0.2, random_state=42)

# Step 4: Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 5: Make predictions and calculate error
y_predict = model.predict(X_test)

# Calculate root mean squared error (RMSE)
val_error = mean_squared_error(y_test, y_predict, squared=False)
val_error
```

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, valid_scores = learning_curve(
    LinearRegression(), housing_features_prepared, housing_target, train_sizes=np.linspace(0.01, 1.0, 40), cv=5,
    scoring="neg_root_mean_squared_error")
train_errors = -train_scores.mean(axis=1)
valid_errors = -valid_scores.mean(axis=1)

plt.figure(figsize=(6, 4))  # extra code â€“ not needed, just formatting
plt.plot(train_sizes, train_errors, "r-+", linewidth=2, label="train")
plt.plot(train_sizes, valid_errors, "b-", linewidth=3, label="valid")

# extra code â€“ beautifies and saves Figure 4â€“15
plt.xlabel("Training set size")
plt.ylabel("RMSE")
plt.grid()
plt.legend(loc="upper right")
#plt.axis([0, 100, 0, 250])

plt.title("Linear")
plt.show()
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from scikeras.wrappers import KerasClassifier
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(housing_features_prepared, housing_target, test_size=0.2, random_state=42)
```

```python
# Install TensorBoard profile
%pip install -q -U tensorboard-plugin-profile
```

```python
from keras.callbacks import TensorBoard
import datetime
from time import strftime

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

def get_run_logdir(root_logdir="my_logs"):
    return Path(root_logdir) / strftime("run_%Y_%m_%d_%H_%M_%S")

run_logdir =  get_run_logdir()
```

```python
# Creating a sequential model
model = Sequential()

# Adding input layer and first hidden layer
model.add(tf.keras.layers.Input(shape=[X_train.shape[1]]))

# Adding first hidden layer
model.add(Dense(10000, activation='relu', kernel_initializer="he_normal"))

# Adding the output layer
model.add(Dense(1, activation='linear'))

# Compiling the model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae'])

# Setting up TensorBoard logs
tensorboard_cb = tf.keras.callbacks.TensorBoard(run_logdir,
                                             profile_batch=(100,200))

# Training the model
model.fit(X_train.toarray(), y_train, epochs=1500, verbose=1, callbacks=[tensorboard_cb])
```

```python
# Making predictions
predictions = model.predict(X_test)

# Evaluating the model
neural_val_error = mean_squared_error(y_test, predictions, squared=False)
neural_val_error
```
