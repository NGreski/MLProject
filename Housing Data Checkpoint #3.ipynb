```python
#Noah Greski - Testing Data ML project

#imports
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
#read in data
housing_data = pd.read_csv(Path("housing data.csv"))
housing_data
```

```python
housing_data.head()
```

```python
housing_data.info()
```

```python
#check for missing values
housing_data.isnull().sum()
```

```python
#create dataframe with only numerical data
housing_num = housing_data.select_dtypes(include=[np.number])
housing_num.head()
```

```python
#visualize the data
g = sns.PairGrid(housing_num)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
#visual the housing prices
plt.xlim = (0, 1000000)
housing_data.sort_values(by = 'price', ascending = False)
```

```python
#housing_data['price'].hist(bins =10)
plt.figure(figsize=(10, 6))
plt.hist(housing_data['price'], bins=20, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histogram of Prices')
plt.xlabel('Price (Tens of Millions)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Separate the houses by price over/under $5 million
condition_value = 1000000  # Million dollar homes
condition_column = 'price'  # Sort by price

# Split based on condition
over = housing_data[housing_data['price'] >= condition_value]
under = housing_data[housing_data['price'] < condition_value]

print("Homes Over $1m: ", len(over), "\nHomes Under $1m: ", len(under))
```

```python
# Revisualize for o/u $1m homes
plt.figure(figsize=(10, 6))
plt.hist(under['price'], bins=10, color='blue', edgecolor='black') # <$1m
plt.title('Histogram of Home Prices Under $1m')
plt.xlabel('Price (Hundreds of Thousands)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(over['price'], bins=50, color='red', edgecolor='black') # >=$1m
plt.title('Histogram of Home Prices Under $1m')
plt.xlabel('Price (Tens of Millions)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
